```python
# 3 check for outliers 
def detect_outliers_all_df(df):
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
    outlier_counts = {}
    
    for column in numerical_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_counts[f'{column}_Outlier'] = column_outliers.shape[0]
    
    return pd.Series(outlier_counts)

outliers_summary_transaction = detect_outliers_all_df(Transaction_df)
outliers_summary_customer = detect_outliers_all_df(Customer_df)
outliers_summary_city = detect_outliers_all_df(City_df)
outliers_summary_cab = detect_outliers_all_df(Cab_df)

print("Summary of outliers in each column for Transaction_df:")
print(outliers_summary_transaction)
print("\nSummary of outliers in each column for Customer_df:")
print(outliers_summary_customer)
print("\nSummary of outliers in each column for City_df:")
print(outliers_summary_city)
print("\nSummary of outliers in each column for Cab_df:")
print(outliers_summary_cab)
```

## Methodology for Outlier Detection
The outlier detection was conducted using the Interquartile Range (IQR) method. For each numerical column, outliers were defined as data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR. This method is robust and commonly used in statistical analysis to identify values that deviate significantly from the central tendency of a data distribution.

## Hypothesis 1: Seasonality in Cab Usage


```python

# Convert 'Date of Travel' to datetime
master_df['Date of Travel'] = pd.to_datetime(master_df['Date of Travel'])

# Extract month from 'Date of Travel'
master_df['Month'] = master_df['Date of Travel'].dt.month

# Group by month and count the number of transactions
monthly_rides = master_df.groupby('Month').size()

# Plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
monthly_rides.plot(kind='bar')
plt.title('Monthly Cab Rides')
plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

## Hypothesis 1: Seasonality in Cab Usage

The bar chart representing the monthly distribution of cab rides shows a distinctive trend. There is a noticeable increase in the number of cab rides towards the year's end, with December having the highest count. This pattern suggests a potential seasonality in cab usage, which could be influenced by various factors:

- **Holiday Season**: The surge in cab rides from September to December aligns with the holiday season. Increased travel, shopping, and social gatherings during this time may contribute to the higher demand for cab services.

- **Weather Conditions**: Weather changes in the later months of the year might also affect the preference for cab rides over other modes of transportation.

- **Tourism and Events**: The end of the year is often packed with events and tourism activities, which could lead to an uptick in cab usage, especially in cities known for holiday festivities.

Further analysis could involve cross-referencing this data with specific city events, weather patterns, and tourism statistics to validate the observed seasonal trend.

```python
# Calculate margin for each ride
master_df['Margin'] = master_df['Price Charged'] - master_df['Cost of Trip']

# Group by date and calculate average margin and number of rides
daily_analysis = master_df.groupby(master_df['Date of Travel'].dt.date).agg({'Margin':'mean', 'Transaction ID':'count'}).rename(columns={'Transaction ID':'Number of Rides'})

# Plotting the relationship
plt.figure(figsize=(10, 6))
plt.scatter(daily_analysis['Number of Rides'], daily_analysis['Margin'])
plt.title('Daily Margin vs. Number of Rides')
plt.xlabel('Number of Rides')
plt.ylabel('Average Daily Margin')
plt.show()
```

There does not appear to be a strong, consistent pattern where an increase in the number of rides is associated with an increase or decrease in margin.

- **Potential Outliers**: There are a few data points with a significantly higher average daily margin compared to the rest, which could be outliers or special cases on certain days that warrant further investigation.

## Hypothesis 5: Cab Usage Frequency Correlates with City Population


```python
# Remove commas from the 'Population' column and convert it to numeric
master_df['Population'] = master_df['Population'].str.replace(',', '').astype(float)

#group by 'City' and correlate population with the number of rides
city_ride_frequency = master_df.groupby('City').agg({
    'Transaction ID': 'count',
    'Population': 'first'  # Assuming Population is the same for all rows of the same city
}).rename(columns={'Transaction ID': 'Number of Rides'})

correlation_population_rides = city_ride_frequency.corr().iloc[0,1]

print(f"The correlation between city population and number of rides is: {correlation_population_rides}")
```

## Hypothesis 5: Correlation Between City Population and Cab Usage Frequency

The analysis of the relationship between city population size and the frequency of cab rides yielded a correlation coefficient of approximately 0.802. This suggests a strong positive linear relationship between the two variables. Key interpretations of this result are:

- **Strong Positive Relationship**: A correlation of 0.802 indicates that as the population of a city increases, the number of cab rides also tends to increase. This is consistent with the expectation that larger cities, with more potential customers, would have higher cab usage.

- **Potential Factors**: While a strong correlation is present, it's important to consider underlying factors that might contribute to this relationship. Larger cities may have more developed cab services, higher demand due to urban density, or a greater preference for cab services over other forms of transportation.

- **Implications for Service Expansion**: For cab companies looking to expand their services, targeting larger cities could be a strategic move given the higher usage rates. However, market saturation and competition should also be considered in theseip fully.
